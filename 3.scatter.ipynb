#### scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Scatter plot with Volume and Girth
plt.scatter(base.Girth, base.Volume, color = 'blue', facecolors = 'none', marker = '*')
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

```python
# Line chart with Volume and Girth
plt.plot(base.Girth, base.Volume)
plt.title('Árvores')
plt.xlabel('Volume')
plt.ylabel('Circunferência')
```

```python
# Scatter plot with jitter
#fit_reg  tendency line
sns.regplot(base.Girth, base.Volume, data = base, x_jitter = 0.3, fit_reg = True)
```
